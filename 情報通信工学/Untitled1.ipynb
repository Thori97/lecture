```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.show()
```

```python
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
import numpy as np
```

```python
w = np.linspace(0, 2*np.pi, 100)
```

```python
x = np.cos(w)
y = np.sin(w)
```

```python
ax.set_ylim([-1.2,1.2])
ax.set_xlim([-1.2,1.2])
ax.set_aspect('equal')
ax.scatter(np.cos(np.pi/4), np.sin(np.pi/4), color = 'r', linewidths="1", alpha = 1)
ax.scatter(np.cos(3*np.pi/4), np.sin(3*np.pi/4), color = 'r', linewidths="2", alpha = 1)
ax.scatter(np.cos(5*np.pi/4), np.sin(5*np.pi/4), color = 'r', linewidths="2", alpha = 1)
ax.plot(x, y, color = 'black')
fig
```
