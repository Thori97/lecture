```python
import matplotlib.pyplot as plt
```

```python
import math
import numpy as np
```

```python
SNR = np.linspace(-0.95, 10, 100)
```

```python
CW = []
for s in SNR:
    CW.append(math.log2(1+s))
```

```python
plt.xlim(-2, 10)
plt.xlabel("SNR[dB]")
plt.ylabel("C/W[bit/s/Hz]")
plt.grid()
plt.plot(SNR, CW)
plt.savefig("1.png")
```

```python
EbN0 = np.linspace(0.000001, 6, 100)
CW2 = []
for x in EbN0:
    CW2.append(10*math.log10((2**x-1)/x))
plt.xlabel("Eb/N0[dB]")
plt.ylabel("C/W[bit/s/Hz]")
plt.xlim(-0.2, 3)
plt.ylim(-2, 4)
plt.grid()
plt.plot(EbN0, CW2)
plt.savefig("2.png")
```
